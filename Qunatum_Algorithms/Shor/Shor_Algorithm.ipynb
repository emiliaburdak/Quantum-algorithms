```python
def iqft(n):
    qc = QuantumCircuit(n)
    for target in reversed(range(n)):
        for control in reversed(range(target +1, n)):
            k = control - target + 1
            angle = -2 * np.pi / (2 ** k)
            qc.cp(angle, control, target)
        qc.h(target)
    for i in range(n // 2):
        qc.swap(i, n - i - 1)
    return qc
```
